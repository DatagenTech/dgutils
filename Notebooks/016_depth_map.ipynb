# Depth Map Visualization
The depth is not easy to visualize since the values range from zero (very close to camera) to infinity (background). <br>
 Since we are interested in visualizing the face area only, we'll perform some steps that will let us achieve this purpose.

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import datagen
```

## Creation of a Dataset container

```python
DATASET_PATH= '../resources/faces_1'
datapoint_index = 0
dp = datagen.load(DATASET_PATH)[datapoint_index]
```

## Depth map visualization

We'll normalize and clamp infinite values from the depth map for better visualization

```python
def normalize_depth_map(depth_map):
    depth_map = depth_map.copy()
    depth_map[depth_map > 2**15] = depth_map[depth_map < 2**15].max()
    depth_map -= depth_map.min()
    depth_map /= depth_map.max()
    depth_map = 1 - depth_map
    return depth_map

plt.figure(figsize=(10, 10))
plt.imshow(normalize_depth_map(dp.depth))
plt.xticks([]); plt.yticks([])
```
