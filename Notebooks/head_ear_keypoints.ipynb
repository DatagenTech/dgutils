# Head and Ear Key-Points

# Imports

```python
import datagen
from dghelpers import show_keypoints
```

## Ear Key-points

```python
#load a dataset of 3 environments from Data3 directory
ds = datagen.load('../Data2',)
#select the first scene in the dataset by index
datapoint = ds[4]
ear_key_points = datapoint.keypoints.ear.right.coords_2d
ear_visible_flags = datapoint.keypoints.ear.right.is_visible

#plot the RGB or "visible_spectrum" image along with the 2d keypoints.
show_keypoints(datapoint.visible_spectrum,ear_key_points,ear_visible_flags,'Ear Key-points')
```

## Back of Head Key-points

```python
#plot the back of head key points
#select the second scene in the dataset by index
datapoint = ds[3]
back_head_keypoints = datapoint.keypoints.head.back.coords_2d
back_head_visible_flags = datapoint.keypoints.head.back.is_visible

#plot the RGB or "visible_spectrum" image along with the 2d keypoints.
show_keypoints(datapoint.visible_spectrum,back_head_keypoints,back_head_visible_flags,'Back of Head Key-points')
```
