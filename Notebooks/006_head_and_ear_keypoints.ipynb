# Head and Ear Keypoints

In this notebook we will learn to extract and visualize the keypoints that Datagen provide for the ear and scalp area. These are highly unique annotations, provided specifically by Datagen. To read more about these, please visit:
- https://docs.datagen.tech/en/latest/Modalities/Head%20keypoints.html
- https://docs.datagen.tech/en/latest/Modalities/Ear%20keypoints.html

# Imports

```python
import datagen
from scripts.dghelpers import show_keypoints
```

## Ear Key-points

```python
ds = datagen.load('../resources/faces_2',)
#select the 5th scene in the dataset by index
datapoint = ds[4]
ear_key_points = datapoint.keypoints.ear.right.coords_2d
ear_visible_flags = datapoint.keypoints.ear.right.is_visible

#plot the RGB or "visible_spectrum" image along with the 2d keypoints.
show_keypoints(datapoint.visible_spectrum,ear_key_points,ear_visible_flags,'Ear Key-points')
```

## Back of Head Key-points

```python
#plot the back of head key points
#select the forth scene in the dataset by index
datapoint = ds[3]
back_head_keypoints = datapoint.keypoints.head.back.coords_2d
back_head_visible_flags = datapoint.keypoints.head.back.is_visible

#plot the RGB or "visible_spectrum" image along with the 2d keypoints.
show_keypoints(datapoint.visible_spectrum,back_head_keypoints,back_head_visible_flags,'Back of Head Key-points')
```

```python
datapoint = ds[2]
top_head_keypoints = datapoint.keypoints.head.top.coords_2d
top_head_visible_flags = datapoint.keypoints.head.top.is_visible

#plot the RGB or "visible_spectrum" image along with the 2d keypoints.
show_keypoints(datapoint.visible_spectrum,top_head_keypoints,top_head_visible_flags,'Top of Head Key-Points')
```
