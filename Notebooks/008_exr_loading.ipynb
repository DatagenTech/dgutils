# EXR files

Datagen's data comes fully annotated with visual and textual modalities. Two of these visual modalities are - depth map and normals map. These two modalities are highly useful for 3D face reconstruction, skin tone estimation, lighting estimation, and many other CV tasks. 

<br>

Our depth and normal maps are saved as EXR files - raster images saved in a special format. The OpenEXR format is a high dynamic range image file format, similar to the HDR raster image format.
This format is used to store images in high quality with a great color and luminance range.
(As a comparison, if the depth and normal maps were saved in regular image formats (PNG and JPEG, 8 bits per channel), the values would not be accurate enough and infinite values would not be represented properly.) 

<br>

An EXR file can be natively visualized in MacOS.<br>
This is a short explanation on how to open and visualize it in Python with OpenCV.

## Imports

```python
# OpenEXR is disabled by default in OpenCV. We must explicitly set up the environment variable to be able to use it
import os
os.environ["OPENCV_IO_ENABLE_OPENEXR"]="1"
import cv2
from scripts.dghelpers import imshow
```

We load the image with OpenCV <br>
Since OpenCV loads the image following BGR convention, let's convert it to RGB.

```python
IMG_PATH = '../resources/faces_2/environment_00001/camera_1/normal_maps.exr'
img_bgr = cv2.imread(IMG_PATH, cv2.IMREAD_UNCHANGED)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

imshow(img_rgb)
```
